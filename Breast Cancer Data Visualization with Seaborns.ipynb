```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cancer_data = pd.read_csv('Cancer_Data.csv')
cancer_data = cancer_data.drop(columns = ['Unnamed: 0'])
cancer_data.head()
```

```python
import seaborn as sns
%matplotlib inline
```

# Categorical Data Plots


```python
sns.barplot(x='diagnosis',y='radius_mean',data= cancer_data) #Mean plot
```

```python
sns.barplot(x='diagnosis',y='radius_mean',data= cancer_data, estimator=np.std) #Std PLot
```

```python
sns.countplot(x='diagnosis',data= cancer_data) #Count plot
```

```python
sns.boxplot(x='diagnosis',y='radius_mean', hue = 'children', data= cancer_data,palette='rainbow') #Boxplot of Diagnosis type for radius_mean
```

```python
sns.violinplot(x='diagnosis',y='radius_mean', data= cancer_data,palette='rainbow')
```

```python
sns.stripplot(x='diagnosis',y='radius_mean', data= cancer_data,palette='rainbow')
```

```python
sns.stripplot(x='diagnosis',y='radius_mean', hue = 'children', data= cancer_data, jitter=True)
```

```python
sns.stripplot(x='diagnosis',y='radius_mean', data= cancer_data,jitter=True,hue='children',palette='Set1',split=True)
```

```python
sns.swarmplot(x='diagnosis',y='radius_mean', data= cancer_data, hue='children', palette="Set1", split=True)
```

# Distribution Plot

```python
sns.distplot(cancer_data['radius_mean'])
```

```python
sns.distplot(cancer_data['radius_mean'],kde=False,bins=30)
```

```python
sns.jointplot(x='radius_mean',y='texture_mean',data= cancer_data,kind='scatter')
```

```python
sns.jointplot(x='radius_mean',y='texture_mean',data= cancer_data,kind='hex')
```

```python
sns.jointplot(x='radius_mean',y='texture_mean',data= cancer_data,kind='reg')
```

```python
sns.pairplot(cancer_data, vars = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 
                                  'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 
                                  'symmetry_mean'])
```

```python
sns.pairplot(cancer_data, vars = ['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 
                                  'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 
                                  'symmetry_mean'], hue = 'children', palette="husl")
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(cancer_data.corr())
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(cancer_data.corr(),cmap='coolwarm',annot=True)
```

```python
sns.lmplot(x='radius_mean',y='texture_mean',data= cancer_data ,hue='diagnosis',palette='coolwarm',
          aspect=0.6,size=8)
```
