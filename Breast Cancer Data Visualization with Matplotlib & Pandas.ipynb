```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
cancer_data = pd.read_csv('Cancer_Data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data = cancer_data.drop(columns = ['Unnamed: 0'])
cancer_data.head()
```

```python

x = np.arange(len(cancer_data))

fig = plt.figure(figsize=(10,4), dpi=100)
plt.plot(x,  cancer_data['radius_mean'], 'skyblue')
plt.xlabel('#')
plt.ylabel('Radius Mean')
plt.title('Line Plot for Radius Mean')
plt.show()
```

```python
fig = plt.figure(figsize=(10,4), dpi=100)
plt.scatter(x, cancer_data['radius_mean'])
plt.xlabel('#')
plt.ylabel('Radius Mean')
plt.title('Dot Plot for Radius Mean')
plt.show()
```

```python
fig = plt.figure(figsize=(10,4), dpi=100)
plt.hist(cancer_data['radius_mean'], bins = 50, color = 'red', alpha = 0.5)
plt.xlabel('#')
plt.ylabel('Radius Mean')
plt.title('Histogram Plot for Radius Mean')
plt.show()
```
